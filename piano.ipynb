```python
from music21 import *
bpm = 0
stream_len = [0]
accidental = []
stream_container = []
music = []
for x in sheet:
    if x[0] == 'bpm':
        bpm = int(x[1])
    elif x[0] == '#' or x[0] == 'b':
        for y in x[0:]: 
            if type(y) == float:
                break
            accidental.append(y)
    else :
        for y in x[1:]:
            stream_list = []
            notelist = []
            durationlist = []
            str_pre_duration = 0
            str_duration = 1
            str_note = ''
            if y == '/':
                break
            if ' ' in y:
                for i in y:
                    if i == '>':
                        str_duration = str_duration * 2
                    if i == '<':
                        str_duration = str_duration / 2
                n = note.Rest()
                n.quarterLength = str_duration
                stream_list.append(stream.Part(stream.Measure(n)))
                print('[ ]  [{}]'.format(str_duration))
            else:
                for i in y:
                    if str.isdigit(i):
                        str_note = str_note+i
                        if str_note in accidental:
                            str_note = str_note + accidental[0]
                        notelist.append(str_note)
                        durationlist.append(str_duration)
                        str_note = ''
                    elif i == '>':
                        str_duration = str_duration * 2
                        durationlist[-1] = str_duration
                    elif i == '<':
                        str_duration = str_duration / 2
                        durationlist[-1] = str_duration
                    elif i == '.':
                        temp = 1.5 * str_duration - 0.5 * str_pre_duration
                        str_pre_duration = str_duration
                        str_duration = temp
                        durationlist[-1] = str_duration
                    elif i == '+':
                        volume += 20
                    elif i == '-':
                        volume -= 20
                    elif i == '=':
                        volume += 10
                    elif i == '_':
                        volume -= 10
                    elif i == '(' or i == ')':
                        notelist = notelist[0:int(len(notelist)/2)]
                        str_duration = durationlist[int(len(durationlist)/2-1)] + durationlist[int(len(durationlist)/2)]
                        for x in range(int(len(durationlist)/2)):
                            durationlist[x] = str_duration
                        durationlist = durationlist[0:int(len(durationlist)/2)]
                    else:
                       str_note = str_note+i
                       str_pre_duration = 0
                       str_duration = 1
                       volume = 60
                #stream_list.append(tempo.MetronomeMark(number=bpm))
                for i in notelist:
                    n = note.Note(i)
                    n.quarterLength = durationlist[notelist.index(i)]
                    n.volume.velocity = volume
                    stream_list.append(stream.Part(stream.Measure(n)))
            if notelist:
                print('{}  {}  {}'.format(notelist,durationlist,volume))
            stream_list = stream.Score(stream_list)
            stream_container.append(stream_list)
        stream_len.append(len(stream_container))
music.append(tempo.MetronomeMark(number=bpm))
for x in range(len(stream_len)-1):
    music.append(stream.Part(stream.Measure(stream_container[stream_len[x]:stream_len[x+1]])))
output = stream.Score(music)
output.write('midi', 'output/{}.midi'.format(file_name))
pygame_player = midi.realtime.StreamPlayer(output)
pygame_player.play()


```
